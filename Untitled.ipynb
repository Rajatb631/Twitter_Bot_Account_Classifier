```python
# preprocess.py
import re
import pandas as pd
from sklearn.model_selection import train_test_split

def clean_text(text):
    if pd.isna(text):
        return ""
    # remove byte artifacts and escape sequences
    text = re.sub(r"^b'", "", str(text))
    text = re.sub(r"\\x[0-9A-Fa-f]{2}", " ", text)
    text = text.encode('utf-8', 'ignore').decode('unicode_escape', 'ignore')
    # remove urls and mentions
    text = re.sub(r"http\S+", " ", text)
    text = re.sub(r"@[\w_]+", " ", text)
    # remove weird non-alpha numeric except # and '
    text = re.sub(r"[^A-Za-z0-9#'\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df = pd.read_csv("./data/Twitter_Data.csv")
# basic cleaning
df['tweet_clean'] = df['Tweet_text'].apply(clean_text)
df['desc_clean']  = df['Twitter_User_Description'].fillna("").apply(clean_text)
df['username_clean'] = df['Twitter_Account'].fillna("").astype(str).apply(lambda x: x.replace('_',' '))

# create combined input
def make_input(row):
    parts = []
    if row['username_clean']:
        parts.append(f"username: {row['username_clean']}")
    if row['desc_clean']:
        parts.append(f"bio: {row['desc_clean']}")
    if row['tweet_clean']:
        parts.append(f"tweet: {row['tweet_clean']}")
    return " [SEP] ".join(parts)

df['model_input'] = df.apply(make_input, axis=1)

# split into train/val/test
train, test = train_test_split(df, test_size=0.10, stratify=df['Label'], random_state=42)
train, val = train_test_split(train, test_size=0.1111, stratify=train['Label'], random_state=42) # ~80/10/10

train.to_csv("train.csv", index=False)
val.to_csv("val.csv", index=False)
test.to_csv("test.csv", index=False)

print("Saved train/val/test with sizes:", len(train), len(val), len(test))
```

```python
# train_distilbert_20k.py
import pandas as pd
from datasets import load_dataset
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForSequenceClassification,
    Trainer,
    TrainingArguments,
)
import numpy as np
import evaluate

# Load CSVs
data_files = {
    "train": "train_20k.csv",
    "validation": "val_20k.csv",
    "test": "test_20k.csv",
}
dataset = load_dataset("csv", data_files=data_files)

# Preprocess
def process_batch(batch):
    batch["text"] = batch["model_input"]
    batch["label"] = batch["Label"]
    return batch

dataset = dataset.map(process_batch)

# Tokenizer
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=128,
    )

dataset = dataset.map(tokenize_fn, batched=True)

# Load model
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", num_labels=2
)

# Metrics
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")
precision = evaluate.load("precision")
recall = evaluate.load("recall")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    preds = np.argmax(predictions, axis=1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "f1": f1.compute(predictions=preds, references=labels, average="binary")["f1"],
        "precision": precision.compute(predictions=preds, references=labels, average="binary")["precision"],
        "recall": recall.compute(predictions=preds, references=labels, average="binary")["recall"],
    }

# Training settings
training_args = TrainingArguments(
    output_dir="distilbert-twitter-bot-20k",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    fp16=False,  # Set to True if you have GPU
    report_to="none",
)

# Remove unused columns
cols_to_remove = [
    "Twitter_User_Name", "Twitter_Account", "Twitter_User_Description",
    "Tweet_id", "Tweet_created_at", "Tweet_text", "tweet_clean",
    "desc_clean", "username_clean", "model_input", "Label"
]

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"].remove_columns(cols_to_remove),
    eval_dataset=dataset["validation"].remove_columns(cols_to_remove),
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

# Train
trainer.train()

# Save final model
trainer.save_model("distilbert-twitter-bot-20k-best")
```

```python
from transformers import pipeline
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import numpy as np

model_dir = "distilbert-twitter-bot-20k-best"
clf = pipeline("text-classification", model=model_dir, tokenizer=model_dir, return_all_scores=True)

test = pd.read_csv("test_20k.csv")

def get_prob(text):
    preds = clf(text[:512])
    probs = {d['label']: d['score'] for d in preds[0]}
    return probs.get('LABEL_1', probs.get('POSITIVE', 0.0))

test["prob_human"] = test["model_input"].apply(get_prob)
test["pred_label"] = (test["prob_human"] > 0.5).astype(int)

print(classification_report(test["Label"], test["pred_label"]))
print(confusion_matrix(test["Label"], test["pred_label"]))
```
